# 📚 Step 2: Import Libraries

```python
from transformers import pipeline, set_seed
import random
```

# 🤖 Step 3: Load GPT Model

```python
generator = pipeline('text-generation', model='gpt2')
set_seed(42)  # For reproducibility
```

# 📝 Step 4: Define Prompt Inputs

```python
prompts = [
    "Once upon a time",
    "In a world where AI ruled everything",
    "Deep in the jungle, a sound echoed",
    "The robot opened its eyes for the first time",
    "She woke up and realized she was the last human"
]
```

# 🧠 Step 5: Generate and Display Stories

```python
def generate_text(prompt, max_length=100):
    output = generator(prompt, max_length=max_length, num_return_sequences=1)
    return output[0]['generated_text']

print("📖 Generated Short Stories:\n")

for prompt in prompts:
    print(f"Prompt: {prompt}")
    story = generate_text(prompt)
    print(f"Generated Story:\n{story}\n{'-'*80}\n")
```

# 🎯 Step 6: Experiment With Custom Prompt

```python
custom_prompt = "The spaceship slowly descended onto Mars"
print(f"Your Custom Prompt: {custom_prompt}")
custom_story = generate_text(custom_prompt)
print(f"\nGenerated Story:\n{custom_story}")
```
